# 4 types of collegeboard questions:

**Methods and Control Structures** 
These questions require writing independent methods that utilize loops, conditionals, return statements, and parameters. I practiced this format thoroughly with methods such as findFreeBlock, numberOfLines, and getLines, all of which incorporate nested loops and logical conditions.

**Classes**
This type of FRQ involves creating or completing a class that includes constructors, fields, and various methods. My experience with AppointmentBook and Sign reflects this practice. I constructed full class definitions with constructors, instance variables, and supporting methods.

**Array/ArrayList**
These questions will evaluate my skills in iterating through, modifying, and accessing elements in Arrays or ArrayLists. Beyond the examples listed, I worked through College Board FRQs and review questions focused on tasks like shifting elements, conditional deletion, and frequency counting. I simulated testing conditions by timing myself and scoring with official AP rubrics.

**2D Array**
This section tests proficiency with nested loop patterns and correct indexing within two-dimensional arrays. I strengthened these skills by solving problems such as SeatingChart, and ArrayResizer, all drawn from College Board FRQs and prep resources. These exercises were key for mastering loop nesting and boundary logic.

How I Studied for the AP Computer Science A Exam
As the AP Computer Science A exam approached, I made it a priority to prepare in multiple ways so I could feel confident and ready. Here's a breakdown of how I studied and the resources that helped me the most.

---

**1. Princeton Review Prep Book**
I used the Princeton Review AP Computer Science A Prep Book as my main study guide. Here's what I did with it:
- Read through each chapter thoroughly.
- Highlighted key Java concepts and syntax rules.
- Took handwritten notes to reinforce concepts like inheritance, ArrayLists, recursion, and sorting algorithms.

Completed the practice questions and tests at the end of each unit to test my understanding.

---

**2. Practice FRQs (Free Response Questions)**
I practiced as many FRQs as I could find—both from past AP exams and review books. This helped me:
- Get comfortable with the format of the exam.
- Learn how to approach problems under time pressure.
- Strengthen my ability to write methods that manipulate arrays, ArrayLists, and classes.

I made sure to time myself and grade my responses based on the official College Board rubrics.

---

**3. GeeksforGeeks Deep Dive**

I spent a lot of time on GeeksforGeeks, which is a treasure trove of programming resources. Specifically, I used it to:
- Review Java basics and OOP concepts like inheritance and polymorphism.
- Explore sorting/searching algorithms in more detail.
- Understand how recursion works through visual examples.

Whenever I got stuck or needed a deeper explanation, this site really helped clarify things.

---

**4. Asking Mr. Mortenson**
When in doubt, I asked questions in class or during office hours. Mr. Mortenson was super helpful in explaining:
- How to approach FRQs logically.
- The best way to structure Java methods and classes.
- Which topics to focus on most for the exam.

Don’t underestimate how helpful it is to just ask your teacher. Even a quick 2-minute conversation can clear up a lot of confusion.
